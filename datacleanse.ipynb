```python
#cleaning zipcode Column
removecharc = ['-','[','nan']
#nan = null in pandas
def clean_zip(dop):
    dop = str(dop)
#     if dop.startswith('[') or dop == 'nan':
#         return 'NaN'
    for character in removecharc:
        if character in dop:
            character_index = dop.find(character)
            dop = dop[:character_index]
    return dop
df['Zipcode'] = df['Zipcode'].apply(clean_zip) #clean up the one with - behind

#removing row with blank space in zipcode col
# df['Zipcode'].replace('',np.nan,inplace=True) #replace blank space with Nan
df.dropna(subset=['Zipcode'],inplace = True) # drop all row containing Nan value on col Zipcode
# df.loc['A-3276724'] #empty zipcode
df['Zipcode'].head(15)
# df.loc['A-3269532'] #empty zipcode
```
